```python
import random
import matplotlib.image as mpimg

def view_random_images(target_path, target_class):
    
    target_folder = target_path + "/" + target_class
    
    random_image = random.sample(os.listdir(target_folder),1)
    print(random_image)
    
    # Read image
    image = mpimg.imread(target_folder+"/"+ random_image[0])
    plt.imshow(image)
    plt.title(target_class)
    plt.axis('off')
    plt.show()
    
    return image
```

```python
IMAGE_SIZE= (224, 224)
BATCH_SIZE = 32


print("Training Dataset.....")
train_dir = tf.keras.preprocessing.image_dataset_from_directory(dir_path,
                                                               image_size=IMAGE_SIZE,
                                                               batch_size=BATCH_SIZE,
                                                               label_mode='binary')
                                                            
val_dir_path = "/content/drive/My Drive/xray_dataset_covid19/test"

print("Val Dataset....")
val_dir = tf.keras.preprocessing.image_dataset_from_directory(val_dir_path,
                                                             image_size=IMAGE_SIZE,
                                                             batch_size=BATCH_SIZE,
                                                             label_mode='binary')

test_dir_path = "/content/drive/My Drive/xray_dataset_covid19/test"
print("Test Datast...")
test_dir = tf.keras.preprocessing.image_dataset_from_directory(test_dir_path,
                                                              image_size=IMAGE_SIZE,
                                                              batch_size=BATCH_SIZE,
                                                              label_mode='binary')
```

```python
# Check the Class names
class_names = train_dir.class_names
class_names
```

```python
# Checking the images and labels in train_dir
for images, labels in train_dir.take(1):
    images = images[0]/255.
    print(images)
```

```python
# Pretech to our required directories
train_ds = train_dir.prefetch(tf.data.AUTOTUNE)
val_ds = val_dir.prefetch(tf.data.AUTOTUNE)
test_ds = test_dir.prefetch(tf.data.AUTOTUNE)
```

```python
train_ds
```

```python
# Early Stopping Callbacks
early_stopping = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',
                                                 patience=3)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout

# Baseline model
model_0 = Sequential([
    Conv2D(16, 3, activation='relu', input_shape=(224,224,3)),
    MaxPool2D(),
    Dropout(0.5),
    Conv2D(16, 3, activation='relu'),
    MaxPool2D(),
    Dropout(0.5),
    Flatten(),
    Dense(1, activation='sigmoid')
])

# Compile the model
model_0.compile(loss='binary_crossentropy',
               optimizer='adam',
               metrics=['accuracy'])

# Get the summary
model_0.summary()
```

```python
# fit the model
history_model_0 = model_0.fit(train_ds,
                           epochs=5,
                           validation_data=val_ds,
                           callbacks=[early_stopping])
```

```python
# Plot the loss curves
def plot_loss_curves(history):
    
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)


    plt.plot(epochs, acc,  label='Training acc')
    plt.plot(epochs, val_acc,  label='Validation acc')
    plt.title('Training and validation accuracy')
    plt.legend()
    plt.figure()

    plt.plot(epochs, loss,  label='Training loss')
    plt.plot(epochs, val_loss, label='Validation loss')
    plt.title('Training and validation loss')
    plt.legend()
    plt.show()
```

```python
plot_loss_curves(history_model_0)
```

```python
model_1 = Sequential([
    Conv2D(16, 3, activation='relu', input_shape=(224,224,3)),
    MaxPool2D(),
    Dropout(0.5),
    Conv2D(16,3,activation='relu'),
    MaxPool2D(),
    Dropout(0.5),
    Flatten(),
    Dense(1, 'sigmoid')
])

# compile
model_1.compile(loss='binary_crossentropy',
               optimizer='adam',
               metrics=['accuracy'])

# Get summary

model_1.summary()
```

```python
# use mobilnet model
baseline_model = tf.keras.applications.mobilenet.MobileNet(include_top=False)
baseline_model.training = False

# Input Shape
inputs = tf.keras.Input(shape=(224,224,3), name='Input_shape')

# Pass Input shape to our Baseline_model
x = baseline_model(inputs)

# Resscaling
x = tf.keras.layers.experimental.preprocessing.Rescaling(1/255.)(x)

# Dropout
x = tf.keras.layers.Dropout(0.5)(x)

# GlobalAveragePooling2D
x = tf.keras.layers.GlobalAveragePooling2D(name='global_average_pooling_2d')(x)

# Dense layer
outputs = tf.keras.layers.Dense(1, activation='sigmoid', name='outputs')(x)

# Build model
model_2 = tf.keras.Model(inputs, outputs)

# compile the model
model_2.compile(loss=tf.keras.losses.BinaryCrossentropy(),
               optimizer=tf.keras.optimizers.Adam(),
               metrics=['accuracy'])

# summary
model_2.summary()
```

```python
history_model_2 = model_2.fit(train_ds,
                             epochs=5,
                             validation_data=val_ds,
                             callbacks=[early_stopping])
```

```python
plot_loss_curves(history_model_2)
```

```python
# use another model
baseline_model = tf.keras.applications.efficientnet.EfficientNetB0(include_top=False)
baseline_model.trainable = False
inputs = tf.keras.Input(shape=(224,224,3)) # Input shape
x = baseline_model(inputs) # Pass inputs to baseline_model
x = tf.keras.layers.GlobalMaxPooling2D()(x) 
x = tf.keras.layers.Dropout(0.6)(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x) # outputs
model_3 = tf.keras.Model(inputs, outputs) # Build model
model_3.compile(loss='binary_crossentropy',
               optimizer='adam',
               metrics=['accuracy'])
model_3.summary() # summary
```

```python
history_model_3  = model_3.fit(train_ds,
                              epochs=5,
                              validation_data=val_ds,
                              callbacks=[early_stopping])
```

```python
def load_and_prep_image(filename, img_shape=224,scale=True):
    
  # Read in target file (an image)
    img = tf.io.read_file(filename)

  # Decode the read file into a tensor & ensure 3 colour channels 
  # (our model is trained on images with 3 colour channels and sometimes images have 4 colour channels)
    img = tf.image.decode_image(img, channels=3)

  # Resize the image (to the same size our model was trained on)
    img = tf.image.resize(img, size = [img_shape, img_shape])

  # Rescale the image (get all values between 0 and 1)
    img = img/255.
    return img
```

##Make Predictions and Visualize It On Custom Image

```python
def pred_and_plot(model, filename, class_names):
    
    """
    Imports an image located at filename, makes a prediction on it with
    a trained model and plots the image with the predicted class as the title.
    """
    # Import the target image and preprocess it
    img = load_and_prep_image(filename)

    # Make a prediction
    pred = model.predict(tf.expand_dims(img, axis=0))

    # Get the predicted class
    pred_class = class_names[int(tf.round(pred))]

    # Plot the image and predicted class
    plt.imshow(img)
    plt.title(f"Prediction: {pred_class}")
    plt.axis(False);
```
