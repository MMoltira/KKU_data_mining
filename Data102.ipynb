```
plt.scatter(data['x'],data['y'],s=250,c=color_list)
```

## Barchart

เรียกใช้ฟังก์ชันที่จำเป็น Barchart มีอินพุทที่จำเป็นมีทั้งหมด 2ตัว คือ xกับtype โดย x คือ บอกว่าแท่งนี้จะไปตั้งไหนบนแกน  type คือ ความสูง

```
import matplotlib #ส่วนimport คือนำเข้าฟังก์ชันที่จำเป็น
import matplotlib.pyplot as plt 
import numpy as np
```

```
labels = ['G1', 'G2', 'G3', 'G4', 'G5']  #ส่วนที่สองกำหนดโดยการสร้างข้อมูลขึ้นมา ข้อมูลมี3ตัว คือ labels men_means women_means
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
```

```
x = np.arange(len(labels))  # the label locations การกำหนดขนาดให้เท่ากับ0.35
width = 0.35  # the width of the bars
```

#แบบง่าย

```
plt.bar(x - width/2, men_means, width, label='Men') #กราฟแท่งเปรียบเทียบmen กับ women
plt.bar(x + width/2, women_means, width, label='Women')
```

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
```

เพิ่ม set ylabel คือการเซ็ตแกนวายว่าเป็นอะไร

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
```

เพิ่ม set_titel คือการตั้งชื่อกราฟ

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
```

#แบบยาก

เพิ่มสองบรรทัดล่าง คือกkiเปลี่ยนจุดX การเปลี่ยนแกนx

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x) #การชี้ค่าในแนวแกนXที่สร้างมา
ax.set_xticklabels(labels); #เปลี่ยนค่าตัวเลขที่เราตั้งการ ;คือบอกว่าจบแล้ว
```

บรรทัดสุดท้าย legend คือ บอกว่าแต่ละสีมันคืออะไร ชื่อของสีนั้นๆ

```
fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
```

ค่าwidthต้องการเพื่อจะได้รู้ว่าเวลาเลือกแท่งสี่เหลี่ยมเลื่อนไปด้านซ้าย ขวา โดย -ซ้าย +ขวา

#กลับมากราฟง่าย

```
plt.bar(x - width/2, men_means, width, label='Men')
plt.bar(x + width/2, women_means, width, label='Women')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.legend();
```
